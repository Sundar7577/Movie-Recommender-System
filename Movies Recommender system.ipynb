```python
import numpy as np 
import pandas as pd
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
#to merge the two dataframe 
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
#from the merged table we only need some of fews colums that will help in our model ,we will only  extract them and put it in dataframe and left others
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#to check is there any null values in data or not we have to write followinng code
movies.isnull().sum()
```

```python
#we see that three movies doesnot have the overview .since overview is important to create tags and 3 missing values will not effect much so we will drop them
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
#To check the duplicate rows
movies.duplicated().sum()
```

```python
#Extracting rows using Pandas .iloc[]
movies.iloc[0].genres
```

```python
#function to extract the name from each dictionary
#ast module and eval function
def extract(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
movies.iloc[0]
```

```python
def cast_extract(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(cast_extract)
```

```python
def crew_extract(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
            
            
```

```python
movies['crew'] = movies['crew'].apply(crew_extract)
```

```python
movies.head()
```

```python
#now the data in the overview is in string so we have to convert it also in list so that it is easy to concatenate it with other 
#using lamda function
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#to remove the space between the first name or last name or similar typr of two words with same meaning meaning or indications
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
df1 = movies[['movie_id','title','tags']]
```

```python
# to convert the list into strings of the tags colun we use
df1['tags'] = df1['tags'].apply(lambda x:" ".join(x))
```

```python
df1['tags'] = df1['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append( ps.stem(i)) 
    return " ".join(y)
```

```python
df1['tags'] = df1['tags'].apply(stem)
```

```python
# text vectorization using the python library
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df1['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def movies_recommender(movie):
    movie_index = df1[df1['title'] == movie].index[0]
    distances = similarity[ movie_index]
    listof_movies = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in listof_movies:
        print(df1.iloc[i[0]].title)
```

```python
movies_recommender('Inception')
```
